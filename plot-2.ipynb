```python
"""
Plot the compare between approach 1 and approach 2
"""

import matplotlib.pyplot as plt
import numpy as np

# Data from Table 1
as_values_table1 = np.array([
    [21.9872, 27.5674, 25.6884, 24.3844],
    [22.9068, 26.9682, 24.935, 24.6876],
    [24.2726, 28.4864, 29.3974, 26.6778]
])

# Data from Table 2
as_values_table2 = np.array([
    [23.009, 21.9244, 25.2818, 20.9052],
    [21.9954, 23.9964, 30.0862, 20.516],
    [24.76, 29.5632, 27.919, 20.516]
])

predictors = ['THR', 'APS', 'SAPS', 'RAPS']

fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.2
index = np.arange(len(predictors))

for i in range(as_values_table1.shape[0]):
    x = ''
    if i == 0: 
        x = 'Split'
    elif i == 1:
        x = 'Cluster'
    else: x = 'ClassWise'
    ax.bar(index + i * bar_width, as_values_table1[i], bar_width, label=f'Approach 1 - {x}', alpha=0.7)
    ax.bar(index + i * bar_width, as_values_table2[i], bar_width, label=f'Approach 2 - {x}', alpha=0.7, hatch='//')

ax.set_xlabel('Score Functions')
ax.set_ylabel('Average Size')
ax.set_title('AS Values Comparison between Approach 1 and Approach 2')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(predictors)
ax.legend()
plt.savefig('AS_Values_Comparison.png',dpi=300)
plt.show()
```

```python
"""
Plot the compare between Scores, predictors.
"""
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
predictors = ['THR', 'APS', 'SAPS', 'RAPS']
as_values = np.array([
    [21.9872, 22.9068, 24.2726],
    [27.5674, 26.9682, 28.4864],
    [25.6884, 24.935, 29.3974],
    [24.3844, 24.6876, 26.6778]
])

# Plotting the "as" values for each predictor
fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.2
index = np.arange(len(predictors))

for i in range(as_values.shape[1]):
    x = ''
    if i == 0: 
        x = 'Split'
    elif i == 1:
        x = 'Cluster'
    else: x = 'ClassWise'
    ax.bar(index + i * bar_width, as_values[:, i], bar_width, label= x)

ax.set_xlabel('Score Functions')
ax.set_ylabel('AS Values')
ax.set_title('AS Values for Different Predictors')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(predictors)
ax.legend()

plt.savefig('AS_Values_Predictors.png', dpi=300)
plt.show()
```
